```python
year_input = (input("Enter a year"))

if not year_input.isdigit():
    raise ValueError("Year must be an integer.")

year = int(year_input)

if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print(f"True if the year is a leap year")
else:
    print(f"False otherwise")
```

```python
n = int(input("Enter a number: "))

if n % 2 == 1:
    print("Weird")
elif n % 2 == 0 and 2 <= n <= 5:
    print("Not Weird")
elif n % 2 == 0 and 6 <= n <=20:
    print("Wierd")
elif n % 2 == 0 and n > 20:
    print("Not Weird")

```

```python
a = int(input("Enter a: "))
b = int(input("Enter b: "))

if a > b:
    a, b = b, a

if a % 2 == 0:
    first_even = a
else:
    first_even = a + 1

if b % 2 == 0:
    last_even = b
else:
    last_even = b - 1

even_numbers = list(range(first_even, last_even + 1, 2))
print(even_numbers)
```
